```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mp
%matplotlib inline
```

```python
salary_data= pd.read_csv('../input/20112018-salaries-for-san-francisco/Total.csv')
```

```python
def get_employee_details(name):
    try:
        # Filtering the employee details
        employee = salary_data[salary_data['Name'].str.lower() == name.lower()]
        if not employee.empty:
            return employee.to_dict('records')[0]
        else:
            return "Employee not found."
    except Exception as e:
        return f"An error occurred: {e}"
```

```python
def safe_get_employee_details(name):
    try:
        if not isinstance(name, str):
            raise ValueError("Input must be a string.")
        
        employee = salary_data[salary_data['Name'].str.lower() == name.lower()]
        if employee.empty:
            return "No employee found with that name."
        return employee.to_dict('records')[0]
    
    except ValueError as ve:
        return f"Value Error: {ve}"
    except Exception as e:
        return f"Unexpected error: {e}"
```

```python
import csv
import os
import zipfile

def export_employee_to_csv(name):
    employee = salary_data[salary_data['Name'].str.lower() == name.lower()]
    if employee.empty:
        print("Employee not found.")
        return

    # Create CSV file
    filename = f"{name}_details.csv"
    employee.to_csv(filename, index=False)

    # Create and save into a zipped folder
    with zipfile.ZipFile('Employee_Profile.zip', 'w') as zipf:
        zipf.write(filename)

    # Optional: Remove the CSV after zipping
    os.remove(filename)
    print(f"{filename} exported and zipped successfully.")
```
